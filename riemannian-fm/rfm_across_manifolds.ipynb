## Experiments

All experiments use `seed=42`. The default configuration lives in [`general_fm.yaml`](configs/experiment/general_fm.yaml); experiment sets are defined in YAML files at [`riemannian-fm/configs/`](riemannian-fm/configs/).

### Configuration

[`general_fm.yaml`](configs/experiment/general_fm.yaml) is the base config, adaptable to any manifold and distribution setup.

**Manifold settings**
- `manifold`: geometry to use (`poincare`, `euclidean`, `sphere`);
- `curvature`: curvature of the manifold;
- `dim`: dimensionality of the space..

**Source and target distributions**
- `x0_dist` / `x1_dist`: distribution type for source and target (`gaussian`, ...);
- `n_samples`: number of samples drawn per distribution;
- `std_x0` / `std_x1`: standard deviation of each Gaussian;
- `mean_x0` / `mean_x1`: mean of each Gaussian, as a list of length `dim`;
- `weights`: mixture weights for a Gaussian mixture. `None` for a single Gaussian, or e.g. `[0.5, 0.5]` for a mixture.

For additional optimization and model parameters, see [`general_fm.yaml`](configs/experiment/general_fm.yaml).
