# Exploring Riemannian Flow Matching Across Manifolds

In this notebook, we study how different geometric priors shape generative coverage, sample quality, and diversity. We address this systematically by comparing RFM across constant-curvature latent geometries: *Euclidean*, *spherical*, and *hyperbolic* manifolds [[1]](#ref1). Starting with a sweep over configurations on toy data, we identify particularly informative or surprising cases for deeper analysis. We may also extend the study to real-world data.

Our main research question is: **How do different Riemannian geometric spaces influence the performance and behaviour of flow matching models?**

## Experiments
*Write down training configurations*

Use `general_fm.yaml` as a base configuration that you can adapt to any manifold and distribution setup.

**Manifold settings**
- `manifold`: the geometry to use (`poincare`, `euclidean`, `sphere`) Default: `poincare`
- `curvature`: curvature of the manifold. Default: `1.0`
- `dim`: dimensionality of the space. Default: `3`

**Source and target distributions**
- `x0_dist` / `x1_dist`: distribution type for source and target (`gaussian`, ...). Default: `gaussian`
- `n_samples`: number of samples drawn per distribution. Default: `20000`
- `std_x0` / `std_x1`: standard deviation of each Gaussian. Default: `0.7`
- `mean_x0` / `mean_x1`: mean of each Gaussian, given as a list with length `dim`. Default: `[0.0, 0.0, 0.0]`
- `weights`: mixture weights for a Gaussian mixture. `None` for a single Gaussian, or a list like `[0.5, 0.5]` for a mixture. Default: `None`

To modify, use command line:

`python train.py experiment=general_fm general.curvature=0.5 general.dim=2`
